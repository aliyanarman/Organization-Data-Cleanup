```python
import pandas as pd
```

```python
df = pd.read_csv('staff sample.csv') 
```

```python
df.sample(20) #getting an idea of data
```

```python

df['Emp ID'].is_unique # making sure employee id appears once
```

```python
df = df.drop_duplicates(subset=['Emp ID']) #dropping duplicate employee id's and rows
```

```python
df['Emp ID'].is_unique # checking whether the transformation has worked
```

```python
df = df.dropna(subset=['Name'])  #removing nulls
```

```python
df['Department'] = df['Department'].replace('???', 'TBA') # replacing inconsistent values in department columm
```

```python
df['Department'].str.strip() != df['Department'] # checking whether any inconsistent spacing exists in department column
```

```python
df # current shape of the data after transformation
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce') #fixing date inconsistency
```

```python
df[["City", "Country"]] = (
    df["Work location"]
    .apply(
        lambda x: ["Remote", "Online"]
        if x.strip().lower() == "remote"                      # slicing the "Work location" column into two columns: city, country
        else [i.strip() for i in x.split(",")]
    )
    .apply(pd.Series)
)

df
```

```python
df['Salary'] = df['Salary'].replace('[\$,]', '', regex=True).astype(float) #converting the data into float
df['Salary'] = df['Salary'].round() #rounding off the salary
df["Salary"] = df["Salary"].map(lambda x: f"${x:,.0f}") #standardizing the column presentation
```
